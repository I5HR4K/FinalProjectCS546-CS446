```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import (
    f1_score, accuracy_score, recall_score, precision_score
)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load train & test sets
train_df = pd.read_csv("kdd_train.csv")
test_df  = pd.read_csv("kdd_test.csv")



train_df.head()
```

```python
target_col = "labels"

# KDD categorical columns (explicit & safe)
categorical_cols = ['protocol_type', 'service', 'flag']

# Everything else except target = numeric
numeric_cols = [col for col in train_df.columns
                if col not in categorical_cols and col != target_col]

print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols[:5], "... total:", len(numeric_cols))
```

```python
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
    ('num', StandardScaler(), numeric_cols)
])
```

```python
base_tree = DecisionTreeClassifier(
    criterion="gini",
    max_depth=None,
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight="balanced"
)
```

```python
# Bagging Classifier
model = BaggingClassifier(
    estimator=base_tree,  # modern sklearn syntax
    n_estimators=150,
    max_samples=0.8,
    max_features=1.0,
    bootstrap=True,
    n_jobs=-1,
    random_state=42
)

```

```python
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('classifier', model)
])
```

```python
#train
pipeline.fit(train_df.drop(target_col, axis=1), train_df[target_col])
#test
y_pred = pipeline.predict(test_df.drop(target_col, axis=1))
#evaluate
accuracy = accuracy_score(test_df[target_col], y_pred)
f1 = f1_score(test_df[target_col], y_pred, average='weighted')
recall = recall_score(test_df[target_col], y_pred, average='weighted')
precision = precision_score(test_df[target_col], y_pred, average='weighted')

print(f"Accuracy:  {accuracy:.4f}")
print(f"F1 Score:  {f1:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"Precision: {precision:.4f}")
```

```python
# Apply style
sns.set_style("whitegrid")
sns.set_context("paper")

# Compute metrics
accuracy = accuracy_score(test_df['labels'], y_pred)
f1 = f1_score(test_df['labels'], y_pred, average='weighted')   # weighted F1 for multiclass


# Prepare DataFrame for plotting
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'F1 Score'],
    'Value': [accuracy, f1]
})

# Plot bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x='Metric', y='Value', data=metrics_df)
plt.title("Classification Metrics")
plt.ylim(0, 1)
plt.ylabel("Score")
plt.show()
```
